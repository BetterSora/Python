# 线性核SVM

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sio
%matplotlib inline
```

```python
# 读取数据
load_mat = sio.loadmat('ex6data1.mat')
X = load_mat['X']
y = load_mat['y']
```

```python
print(X.shape, y.shape)
```

```python
# 可视化数据
pos = np.where(y==1)[0]
neg = np.where(y==0)[0]
#plt.scatter(X[:,0], X[:,1],c=y, edgecolors='black')
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='yellow', edgecolors='black')
plt.scatter(X[neg, 0], X[neg, 1], marker='o', c='blue', edgecolors='black')
plt.show()
```

```python
# C=1的情况
clf = SVC(C=1, kernel='linear')
clf.fit(X, y)
theta = clf.coef_
b = clf.intercept_
xx = np.linspace(X[:, 0].min(), X[:, 1].max(), 100)
yy = -(theta[0, 0] * xx + b) / theta[0, 1]
pos = np.where(y==1)[0]
neg = np.where(y==0)[0]
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='yellow', edgecolors='black')
plt.scatter(X[neg, 0], X[neg, 1], marker='o', c='blue', edgecolors='black')
plt.plot(xx, yy)
plt.show()
```

```python
# C=1000的情况
clf = SVC(C=1000, kernel='linear')
clf.fit(X, y)

theta = clf.coef_
b = clf.intercept_

xx = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
yy = -(theta[0, 0] * xx + b) / theta[0, 1];
pos = np.where(y==1)[0]
neg = np.where(y==0)[0]
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='yellow', edgecolors='black')
plt.scatter(X[neg, 0], X[neg, 1], marker='o', c='blue', edgecolors='black')
plt.plot(xx, yy)
plt.show()
```

# 高斯核SVM

```python
# 读取数据
load_mat = sio.loadmat('ex6data2.mat')
X = load_mat['X']
y = load_mat['y']

# 可视化数据
pos = np.where(y==1)[0]
neg = np.where(y==0)[0]
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='yellow', edgecolors='black')
plt.scatter(X[neg, 0], X[neg, 1], marker='o', c='blue', edgecolors='black')
plt.show()
```

```python
# c和gamma越大，bias越小variance越大
clf = SVC(C=1, kernel='rbf', gamma=200)
clf.fit(X,y)

pos = np.where(y==1)[0]
neg = np.where(y==0)[0]
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='yellow', edgecolors='black')
plt.scatter(X[neg, 0], X[neg, 1], marker='o', c='blue', edgecolors='black')

x0 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
x1 = np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
xx, yy = np.meshgrid(x0, x1)
h = clf.predict(np.c_[xx.ravel(), yy.ravel()])
h = h.reshape(xx.shape)
plt.contour(xx, yy, h, [0.5], linewidths=1, colors='r')
plt.show()
```
